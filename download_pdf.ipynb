```python
# Creando WebDriver para la busqueda con Chrome
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
thesis_data = pd.read_parquet("data/thesis.parquet")
thesis_data = thesis_data.query('estado == "Acceso abierto"').reset_index(drop=True)
thesis_data
```

```python
def clean_title(title):
    pattern = r"[" + string.punctuation + "]"
    new_title = re.sub(pattern, "", title)

    return new_title
```

```python
for index, row in thesis_data.iterrows():
    try:
        title = clean_title(row["titulo"])
        url = row["url"]
        PATH = f"data/san marcos/{row['id']}.pdf"

        if not os.path.isfile(PATH):
            driver.get(url)
            pdf_link = driver.find_element(
                By.XPATH,
                '//*[@id="aspect_artifactbrowser_ItemViewer_div_item-view"]/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/a',
            ).get_attribute("href")

            print(row["id"], pdf_link)

            response = requests.get(pdf_link, verify=False)
            open(PATH, "wb").write(response.content)

            # if index % 3 == 0 or index % 7 == 0:
            #     driver.get(random.choice(redes_sociales))
            #     time.sleep(random.randint(5, 12))
            # else:
            #     time.sleep(4)
    except:
        print("Error", title)
```
